**(b) Removing the units from Mileage, Engine, Power and New_Price and converting them into float**

```python
df= pd.DataFrame(markertingdata)
df['Mileage'] = df['Mileage'].str.extract('(\d+\.\d+)').astype(float)

df['Engine'] = df['Engine'].str.replace(' CC', '').astype(float)

df['Power'] = df['Power'].str.extract('(\d+\.\d+)').astype(float)

df['New_Price'] = df['New_Price'].str.extract('(\d+\.\d+)').astype(float)

df.head(5)
```

##### **(a) Filling in missing values in Mileage, Engine, Power,Seats**

```python
df['Mileage'].fillna(df['Mileage'].mean(),inplace=True)
df['Engine'].fillna(df['Engine'].mean(),inplace=True)
df['Power'].fillna(df['Power'].mean(),inplace=True)

df['Seats'].fillna(df['Seats'].mode()[0],inplace=True)

df.drop('New_Price', axis=1, inplace=True)#axis=1 indicates to drop a column
```

```python
output_csv_file = 'Preprocessed_data.csv'
df.to_csv(output_csv_file, index=False)
```

###### **(c) To work with our categorical variables, we are converting them into numerical data using one-hot encoding.**

```python
df1['Fuel_Type'].replace({'Diesel': 0, 'Petrol': 1, 'Electric': 2}, inplace=True)
df1['Transmission'].replace({'Manual': 0, 'Automatic': 1}, inplace=True)
df1['Owner_Type'].replace({'First': 1, 'Second': 2, 'Third': 3, 'Fourth & Above': 4}, inplace=True)
```

##### e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset.

```python
selected_columns = df1[['Name', 'Mileage', 'Year', 'Engine', 'Price','Location']]
selected_columns
```

```python
df1['Mileage_per_Year'] = df1['Mileage'] / df1['current_age']
df1['Mileage_per_Year']
```

```python
# Filtering based on a condition
filtered_data = renamed_df1[df1['Fuel_Type'] == 1]
filtered_data
```

```python
# Arranging data
arranged_data = filtered_data.sort_values(by='Price')
arranged_data.head(7)
```
